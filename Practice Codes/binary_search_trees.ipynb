```python

class TreeNode(object):

    def __init__(self, value, left_child = None, right_child = None):
        self.val = value
        self.left = left_child
        self.right = right_child

    
node1 = TreeNode(25)
node2 = TreeNode(75)
root = TreeNode(50, 25, 75)
```

```python

def search_node(searchValue, node):

    current_node = node

    if current_node.val == searchValue or current_node == None:
        return current_node

    elif searchValue < current_node.val:
        return search_node(searchValue, current_node.left)
    
    return search_node(searchValue, current_node.right)



def insert_node(value, node):  # No return keyword for insertion operation.

    current_node = node 

    if current_node.val > value:
        if current_node.left == None:
            current_node.left = TreeNode(value)
        else:
            insert_node(value, current_node.left)
    
    elif current_node.val < value:
        if current_node.right == None:
            current_node.right = TreeNode(value)
        else:
            insert_node(value, current_node.right)

def delete_node(value, node):

    if node == None:
        return None

    elif value < node.val:
        node.left = delete_node(value, node.left)
        return node 
    
    elif value > node.val:
        node.right = delete_node(value, node.right)

    else:
        if not node.left or not node.right: return None
        elif not node.left and node.right: return node.right
        elif not node.right and node.left: return node.left
        else:
            pass

def inorder_traversal(node):
    #Base case
    if not node:
        return 
    # Beautiful hack
    inorder_traversal(node.left)
    print(node.val)
    inorder_traversal(node.right)


    


            


```
