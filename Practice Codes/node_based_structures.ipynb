```python
'''
    Implementing a Queue using Doubly Linked List
'''

class Node(object):

    def __init__(self, value):
        self.val = value
        self.next = None
        self.prev = None

class DoublyLinkedList(object):

    def __init__(self, first_node = None, last_node = None):
        self.first = first_node
        self.last = last_node

    
    def insert_at_end(self, value):
        node = Node(value)

        if not self.first:
            self.first = node
            self.last = node
        
        else:
            self.last.next = node
            node.prev = self.last

            self.last = node
    
    def remove_from_front(self):
        del_node = self.first
        self.first = self.first.next

        return del_node


class Queue(object):

    def __init__(self):
        self.data = DoublyLinkedList()

    def enqueue(self, value):
        self.data.insert_at_end(value)

    def dequeue(self):
        return self.data.remove_from_front()

    def read(self):
        if not self.first:
            return None
        return self.first








```
